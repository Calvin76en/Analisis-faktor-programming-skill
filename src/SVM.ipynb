# SVM dengan metode K-Fold cross validation

### Import Library

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, classification_report, confusion_matrix
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

### Membaca Dataset

```python
df = pd.read_csv(r'C:\Users\Asus\Machine Learning\Projek PM\Analisis-faktor-programming-skill-main\Analisis-faktor-programming-skill-main\data\preprocessing1.csv', sep=',')
```

```python
df.head()
```

### Membagi Fitur dan Target

```python
# Pemisahan X (fitur) dan y (target)
X = df.drop(columns=['Target_Category'])  # Ganti data dengan df
y = df['Target_Category']  # Ganti data dengan df

# Pisahkan data 80% untuk train dan 20% untuk test
X_train_full, X_test, y_train_full, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

###  Stratified K-Fold Cross-Validation

```python
# Setup StratifiedKFold untuk cross-validation pada data train
skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
```

### Inisialisasi Model SVM

```python
# Inisialisasi model SVM tanpa tuning
svm_model = SVC(kernel='rbf', C=1, gamma=0.01, random_state=42)
```

### Cross Validation

```python
# Daftar untuk menyimpan skor sebelum tuning
accuracy_scores_train = []
f1_scores_train = []
accuracy_scores_val = []
f1_scores_val = []

# Cross-validation pada data train untuk model tanpa hyperparameter tuning
for fold, (train_index, val_index) in enumerate(skf.split(X_train_full, y_train_full), 1):
    print(f"Processing fold {fold}...")
    
    # Membagi data train menjadi train dan validation
    X_train, X_val = X_train_full.iloc[train_index], X_train_full.iloc[val_index]
    y_train, y_val = y_train_full.iloc[train_index], y_train_full.iloc[val_index]
    
    # Latih model SVM tanpa tuning
    svm_model = SVC(kernel='rbf', C=1, gamma=0.01, random_state=42)
    svm_model.fit(X_train, y_train)
    
    # Prediksi pada training dan validation set
    y_train_pred = svm_model.predict(X_train)
    y_val_pred = svm_model.predict(X_val)
    
    # Evaluasi metrik pada training set
    accuracy_scores_train.append(accuracy_score(y_train, y_train_pred))
    f1_scores_train.append(f1_score(y_train, y_train_pred, average='weighted'))
    
    # Evaluasi metrik pada validation set
    accuracy_scores_val.append(accuracy_score(y_val, y_val_pred))
    f1_scores_val.append(f1_score(y_val, y_val_pred, average='weighted'))
    
    # Classification report dan confusion matrix untuk tiap fold
    print(f"Classification Report (Fold {fold} - Validation - Before Tuning):\n", classification_report(y_val, y_val_pred))
    
    cm = confusion_matrix(y_val, y_val_pred)
    sns.heatmap(cm, annot=True, fmt="d", cmap='Blues', xticklabels=np.unique(y), yticklabels=np.unique(y))
    plt.title(f"Confusion Matrix (Fold {fold} - Validation - Before Tuning)")
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.show()
```

### Rata-rata Skor Sebelum Hyperparameter Tuning

```python
print("\nAverage Scores (Training + Validation - Before Tuning):")
print(f"Training Accuracy: {np.mean(accuracy_scores_train):.4f} ± {np.std(accuracy_scores_train):.4f}")
print(f"Validation Accuracy: {np.mean(accuracy_scores_val):.4f} ± {np.std(accuracy_scores_val):.4f}")
print(f"Training F1-Score: {np.mean(f1_scores_train):.4f} ± {np.std(f1_scores_train):.4f}")
print(f"Validation F1-Score: {np.mean(f1_scores_val):.4f} ± {np.std(f1_scores_val):.4f}")
```

### Evaluasi Data Test Sebelum Tuning

```python
svm_model.fit(X_train_full, y_train_full)
y_test_pred_before_tuning = svm_model.predict(X_test)
print("\nClassification Report on Test Data (Before Tuning):")
print(classification_report(y_test, y_test_pred_before_tuning))

cm_before_tuning = confusion_matrix(y_test, y_test_pred_before_tuning)
sns.heatmap(cm_before_tuning, annot=True, fmt="d", cmap='Blues', xticklabels=np.unique(y), yticklabels=np.unique(y))
plt.title("Confusion Matrix (Test Set - Before Tuning)")
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```

### Hyperparameter Tuning Menggunakan GridSearchCV

```python
# Menyusun parameter grid untuk GridSearchCV
param_grid = {
    'C': [0.1, 1, 10, 100, 1000],
    'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
    'kernel': ['rbf', 'poly', 'linear']
}
# GridSearchCV untuk optimisasi hyperparameter
grid_search = GridSearchCV(estimator=SVC(), param_grid=param_grid, cv=5, n_jobs=-1, verbose=1)
grid_search.fit(X_train_full, y_train_full)

# Menampilkan parameter terbaik dari GridSearchCV
print(f'\nBest parameters from GridSearchCV: {grid_search.best_params_}')
# Menggunakan model terbaik dari GridSearchCV
best_svm_model = grid_search.best_estimator_
```

### Cross-Validation dengan Model Terbaik (Setelah Tuning)

```python
accuracy_scores_train_tune = []
f1_scores_train_tune = []
accuracy_scores_val_tune = []
f1_scores_val_tune = []

for fold, (train_index, val_index) in enumerate(skf.split(X_train_full, y_train_full), 1):
    print(f"Processing fold {fold} (Tuned model)...")
    
    # Membagi data train menjadi train dan validation
    X_train, X_val = X_train_full.iloc[train_index], X_train_full.iloc[val_index]
    y_train, y_val = y_train_full.iloc[train_index], y_train_full.iloc[val_index]
    
    # Latih model terbaik
    best_svm_model.fit(X_train, y_train)
    
    # Prediksi pada training dan validation set
    y_train_pred = best_svm_model.predict(X_train)
    y_val_pred = best_svm_model.predict(X_val)
    
    # Evaluasi metrik pada training set
    accuracy_scores_train_tune.append(accuracy_score(y_train, y_train_pred))
    f1_scores_train_tune.append(f1_score(y_train, y_train_pred, average='weighted'))
    
    # Evaluasi metrik pada validation set
    accuracy_scores_val_tune.append(accuracy_score(y_val, y_val_pred))
    f1_scores_val_tune.append(f1_score(y_val, y_val_pred, average='weighted'))
    
    # Classification report dan confusion matrix untuk tiap fold
    print(f"Classification Report (Fold {fold} - Validation - After Tuning):\n", classification_report(y_val, y_val_pred))
    
    cm_tune = confusion_matrix(y_val, y_val_pred)
    sns.heatmap(cm_tune, annot=True, fmt="d", cmap='Blues', xticklabels=np.unique(y), yticklabels=np.unique(y))
    plt.title(f"Confusion Matrix (Fold {fold} - Validation - After Tuning)")
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.show()
```

```python
# Evaluasi ulang setelah tuning dengan K-Fold Cross-Validation pada bagian train
accuracy_scores_tune = []
f1_scores_tune = []

for fold, (train_index, val_index) in enumerate(skf.split(X_train_full, y_train_full), 1):
    print(f"Processing fold {fold} (Tuned model)...")
    
    # Membagi data train menjadi train dan validation
    X_train, X_val = X_train_full.iloc[train_index], X_train_full.iloc[val_index]
    y_train, y_val = y_train_full.iloc[train_index], y_train_full.iloc[val_index]
    
    # Latih model terbaik
    best_svm_model.fit(X_train, y_train)
    
    # Prediksi pada validation set
    y_pred = best_svm_model.predict(X_val)
    
    # Evaluasi metrik
    accuracy_scores_tune.append(accuracy_score(y_val, y_pred))
    f1_scores_tune.append(f1_score(y_val, y_pred, average='weighted'))
    
    # Menampilkan classification report dan confusion matrix untuk tiap fold
    print(f"Classification Report (Fold {fold} - After Tuning):\n", classification_report(y_val, y_pred))
    
    cm_tune = confusion_matrix(y_val, y_pred)
    sns.heatmap(cm_tune, annot=True, fmt="d", cmap='Blues', xticklabels=np.unique(y), yticklabels=np.unique(y))
    plt.title(f"Confusion Matrix (Fold {fold} - After Tuning)")
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.show()
```

### Rata-rata Skor Setelah Hyperparameter Tuning

```python
print("\nAverage Scores (Training + Validation - After Tuning):")
print(f"Training Accuracy: {np.mean(accuracy_scores_train_tune):.4f} ± {np.std(accuracy_scores_train_tune):.4f}")
print(f"Validation Accuracy: {np.mean(accuracy_scores_val_tune):.4f} ± {np.std(accuracy_scores_val_tune):.4f}")
print(f"Training F1-Score: {np.mean(f1_scores_train_tune):.4f} ± {np.std(f1_scores_train_tune):.4f}")
print(f"Validation F1-Score: {np.mean(f1_scores_val_tune):.4f} ± {np.std(f1_scores_val_tune):.4f}")
```

```python
print("\n--- Perbandingan Baseline vs Tuning (Rata-rata) ---")

# Rata-rata metrik untuk model baseline pada Test Set
test_performance = {
    'accuracy': accuracy_score(y_test, y_test_pred_before_tuning),
    'precision': precision_score(y_test, y_test_pred_before_tuning, average='weighted'),
    'recall': recall_score(y_test, y_test_pred_before_tuning, average='weighted'),
    'f1_score': f1_score(y_test, y_test_pred_before_tuning, average='weighted')
}

print("\nRata-rata Performance Baseline:")
print(f"Accuracy: {test_performance['accuracy']:.4f}")
print(f"Precision: {test_performance['precision']:.4f}")
print(f"Recall: {test_performance['recall']:.4f}")
print(f"F1 Score: {test_performance['f1_score']:.4f}")

# Rata-rata metrik untuk model setelah tuning pada Test Set
tuned_test_performance = {
    'accuracy': accuracy_score(y_test, y_test_pred_after_tuning),
    'precision': precision_score(y_test, y_test_pred_after_tuning, average='weighted'),
    'recall': recall_score(y_test, y_test_pred_after_tuning, average='weighted'),
    'f1_score': f1_score(y_test, y_test_pred_after_tuning, average='weighted')
}

print("\nRata-rata Performance Setelah Tuning:")
print(f"Accuracy: {tuned_test_performance['accuracy']:.4f}")
print(f"Precision: {tuned_test_performance['precision']:.4f}")
print(f"Recall: {tuned_test_performance['recall']:.4f}")
print(f"F1 Score: {tuned_test_performance['f1_score']:.4f}")
```
