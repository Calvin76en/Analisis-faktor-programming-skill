### Melakukan One-Hot Encoding untuk Kolom Non-Ordinal

Untuk kolom yang berisi data bertipe objek atau kategori non-ordinal, seperti nama media sosial,
kita akan menggunakan One-Hot Encoding untuk mengubahnya menjadi variabel numerik yang bisa diproses.


Multi-label Handling:
Kolom dengan nilai seperti Instagram, TikTok, YouTube mengandung beberapa kategori dalam satu sel, yang tidak bisa langsung diproses oleh algoritma identifikasi outlier. Anda perlu memisahkan setiap kategori menjadi fitur individual dengan one-hot encoding.

Numerical Representation:
Identifikasi outlier (misalnya, dengan Z-score atau IQR) biasanya diterapkan pada data numerik. Setelah one-hot encoding, setiap kategori menjadi fitur biner (0 atau 1), memungkinkan deteksi outlier.

# Menangani Outlier

```python
# Menghitung Q1 (25th percentile) dan Q3 (75th percentile)
Q1 = df_subset.quantile(0.25)
Q3 = df_subset.quantile(0.75)

# Menghitung IQR (Interquartile Range)
IQR = Q3 - Q1

# Menentukan batas bawah dan atas untuk mendeteksi outlier
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Kolom independen (semua kolom kecuali kolom terakhir)
independent_columns = df_subset.columns[:-1]

# Kolom dependen (kolom terakhir)
dependent_column = df_subset.columns[-1]

# Menangani outlier untuk kolom independen
for col in independent_columns:
    df_subset[col] = df_subset[col].clip(lower=lower_bound[col], upper=upper_bound[col])

# Menangani outlier untuk kolom dependen secara terpisah
# Menghitung batas bawah dan atas untuk kolom dependen
lower_bound_dep = lower_bound[dependent_column]
upper_bound_dep = upper_bound[dependent_column]

# Pilihan: Clip nilai outlier pada kolom dependen (opsional, tergantung kebutuhan)
df_subset[dependent_column] = df_subset[dependent_column].clip(lower=lower_bound_dep, upper=upper_bound_dep)

# Mengecek hasil setelah mengganti outlier
print(f"Dataset setelah menangani outlier:\n{df_subset.head()}")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Pilih semua fitur numerik dari dataset kecuali target
features = df_selection.drop(columns=['Programming Skill']).columns
target = 'Programming Skill'

# Iterasi melalui setiap fitur
for feature in features:
    X = df_selection[[feature]]  # Fitur (dipilih satu per satu)
    y = df_selection[target]    # Target

    # Model regresi linier
    model = LinearRegression()
    model.fit(X, y)
    y_pred = model.predict(X)

    # R-squared
    r_squared = r2_score(y, y_pred)
    print(f'R-squared untuk {feature}: {r_squared:.4f}')

    # Visualisasi scatter plot dan regresi linier
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=X.squeeze(), y=y, label="Data Asli")
    plt.plot(X, y_pred, color='red', label="Regresi Linier")
    plt.xlabel(feature)
    plt.ylabel(target)
    plt.title(f'Regresi Linier: {feature} vs {target}\nR-squared: {r_squared:.4f}')
    plt.legend()
    plt.show()
```

# SMOTE (Balancing Target Classes)

Mengatasi ketidakseimbangan data target dengan oversampling kelas minoritas menggunakan SMOTE.

```python
# Periksa distribusi kelas target
print(df_selection['Target_Category'].value_counts())
```

```python
df_selection['Target_Category'] = df_selection['Target_Category'].map({'Low': 0, 'Medium': 1, 'High': 2})
```

```python
# Pisahkan fitur dan target, pastikan untuk drop kolom 'Target_Category' dan 'Programming Skill'
X = df_selection.drop(columns=['Target_Category', 'Programming Skill'])  # Fitur (X)
y = df_selection['Target_Category']  # Target variabel (y)

# Periksa distribusi target sebelum SMOTE
print(y.value_counts())
```

```python
from imblearn.over_sampling import SMOTE

# Terapkan SMOTE untuk menangani ketidakseimbangan kelas
smote = SMOTE(random_state=42)

# Resampling fitur dan target menggunakan SMOTE
X_resampled, y_resampled = smote.fit_resample(X, y)

# Periksa distribusi kelas setelah SMOTE
print(pd.Series(y_resampled).value_counts())
```

```python
# Kembalikan target yang sudah di-resample ke dalam bentuk kategori teks
y_resampled = y_resampled.map({0: 'Low', 1: 'Medium', 2: 'High'})

# Periksa distribusi target setelah kembalikan ke kategori
print(pd.Series(y_resampled).value_counts())
```

```python
# Gabungkan fitur dan target
df_resampled = pd.DataFrame(X_resampled, columns=X.columns)  # Data fitur
df_resampled['Target_Category'] = y_resampled  # Data target yang sudah diproses
```
