# --- 1. Import Library yang Dibutuhkan ---

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold, train_test_split, GridSearchCV
from sklearn.metrics import classification_report, accuracy_score, f1_score, precision_score, recall_score, confusion_matrix
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# --- 2. Membaca Dataset ---

```python
df = pd.read_csv('D:/FOLDER MUTIARA/SEMESTER 5/PM/Project/Analisis-faktor-programming-skill/data/preprocessing1.csv', sep=',')
```

# --- 3. Menentukan Fitur dan Target ---

```python
X = df.drop('Target_Category', axis=1)  # Fitur
y = df['Target_Category']  # Target
```

# --- 4. Membagi Data Menjadi Train+Validation dan Test Set ---

```python
X_train_val, X_test, y_train_val, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
print(f"Train+Validation size: {len(X_train_val)}, Test size: {len(X_test)}")
```

# --- 5. Stratified K-Fold Cross Validation ---

```python
skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
```

# Menyiapkan list untuk menyimpan metrik evaluasi

```python
accuracy_scores = []
f1_scores = []
precision_scores = []
recall_scores = []
```

# --- 6. Proses Cross Validation ---

```python
for fold, (train_index, val_index) in enumerate(skf.split(X_train_val, y_train_val), 1):
    # Membagi data ke dalam training dan validation
    X_train, X_val = X_train_val.iloc[train_index], X_train_val.iloc[val_index]
    y_train, y_val = y_train_val.iloc[train_index], y_train_val.iloc[val_index]

    # Membuat dan melatih model RandomForest
    rf = RandomForestClassifier(random_state=42)
    rf.fit(X_train, y_train)

    # Prediksi pada validation set
    y_pred_val = rf.predict(X_val)

    # Menghitung metrik evaluasi untuk validation set
    accuracy = accuracy_score(y_val, y_pred_val)
    precision = precision_score(y_val, y_pred_val, average='weighted')
    recall = recall_score(y_val, y_pred_val, average='weighted')
    f1 = f1_score(y_val, y_pred_val, average='weighted')

    # Menyimpan hasil evaluasi
    accuracy_scores.append(accuracy)
    f1_scores.append(f1)
    precision_scores.append(precision)
    recall_scores.append(recall)

    # Menampilkan hasil evaluasi per fold
    print(f"Fold {fold} - Validation Set Performance:")
    print(f"Accuracy: {accuracy:.4f}, Precision: {precision:.4f}, Recall: {recall:.4f}, F1 Score: {f1:.4f}")
    print("\n")

    # Menampilkan classification report untuk tiap fold
    print(f"Classification Report (Fold {fold} - Validation):\n", classification_report(y_val, y_pred_val))

    # Menampilkan confusion matrix untuk tiap fold
    cm = confusion_matrix(y_val, y_pred_val)
    sns.heatmap(cm, annot=True, fmt="d", cmap='Blues', xticklabels=np.unique(y), yticklabels=np.unique(y))
    plt.title(f"Confusion Matrix (Fold {fold} - Validation)")
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.show()
```

# --- 7. Rata-Rata Metrik untuk Train+Validation Set ---

```python
val_performance = {
    'accuracy': round(np.mean(accuracy_scores), 4),
    'f1_score': round(np.mean(f1_scores), 4),
    'precision': round(np.mean(precision_scores), 4),
    'recall': round(np.mean(recall_scores), 4)
}

print("\nPerformance on Train+Validation Set:")
for metric, value in val_performance.items():
    print(f"{metric.capitalize()}: {value:.4f}")
```

# --- 8. Evaluasi Model pada Test Set ---

```python
rf.fit(X_train_val, y_train_val)  # Melatih ulang pada seluruh Train+Validation
y_pred_test = rf.predict(X_test)

# Menghitung performa model pada test set
test_performance = {
    'accuracy': accuracy_score(y_test, y_pred_test),
    'precision': precision_score(y_test, y_pred_test, average='weighted'),
    'recall': recall_score(y_test, y_pred_test, average='weighted'),
    'f1_score': f1_score(y_test, y_pred_test, average='weighted')
}

print("\nPerformance on Test Set:")
print(f"Accuracy: {test_performance['accuracy']:.4f}")
print(f"Precision: {test_performance['precision']:.4f}")
print(f"Recall: {test_performance['recall']:.4f}")
print(f"F1 Score: {test_performance['f1_score']:.4f}")
```

# --- 9. GridSearchCV untuk Hyperparameter Tuning ---

```python
param_grid = {
    'criterion': ['gini', 'entropy', 'log_loss'],  # Kriteria untuk pemilihan split
    'n_estimators': [7, 100, 150],  # Jumlah pohon dalam hutan
    'max_features': ['sqrt', 'log2'],  # Max features yang dipilih untuk split
    'max_depth': [None, 32, 64, 112, 128],  # Kedalaman maksimum pohon
    'min_samples_split': [2, 4, 6],  # Jumlah sampel minimum untuk split
}

# Inisialisasi GridSearchCV
grid_search = GridSearchCV(
    estimator=RandomForestClassifier(random_state=42),
    param_grid=param_grid,
    scoring='accuracy',
    cv=5,
    verbose=2,
    n_jobs=-1
)

# Melakukan pencarian hyperparameter
grid_search.fit(X_train_val, y_train_val)

# Menampilkan hasil best hyperparameters
print("\nBest Hyperparameters:", grid_search.best_params_)
best_model = grid_search.best_estimator_
```

# --- 10. Evaluasi Model Setelah Tuning pada Train+Validation Set ---

```python
accuracy_scores_tuned = []
f1_scores_tuned = []
precision_scores_tuned = []
recall_scores_tuned = []

# Proses Cross Validation pada model setelah tuning
for fold, (train_index, val_index) in enumerate(skf.split(X_train_val, y_train_val), 1):
    X_train, X_val = X_train_val.iloc[train_index], X_train_val.iloc[val_index]
    y_train, y_val = y_train_val.iloc[train_index], y_train_val.iloc[val_index]

    # Latih model setelah tuning
    best_model.fit(X_train, y_train)

    # Prediksi pada validation set
    y_pred_val_tuned = best_model.predict(X_val)

    # Evaluasi pada validation set
    accuracy = accuracy_score(y_val, y_pred_val_tuned)
    precision = precision_score(y_val, y_pred_val_tuned, average='weighted')
    recall = recall_score(y_val, y_pred_val_tuned, average='weighted')
    f1 = f1_score(y_val, y_pred_val_tuned, average='weighted')

    # Menyimpan hasil evaluasi
    accuracy_scores_tuned.append(accuracy)
    f1_scores_tuned.append(f1)
    precision_scores_tuned.append(precision)
    recall_scores_tuned.append(recall)

    # Menampilkan hasil evaluasi per fold setelah tuning
    print(f"Fold {fold} - Validation Set Performance After Tuning:")
    print(f"Accuracy: {accuracy:.4f}, Precision: {precision:.4f}, Recall: {recall:.4f}, F1 Score: {f1:.4f}")
    print("\n")

    # Menampilkan classification report untuk tiap fold setelah tuning
    print(f"Classification Report (Fold {fold} - Validation After Tuning):\n", classification_report(y_val, y_pred_val_tuned))

    # Menampilkan confusion matrix untuk tiap fold setelah tuning
    cm_tuned = confusion_matrix(y_val, y_pred_val_tuned)
    sns.heatmap(cm_tuned, annot=True, fmt="d", cmap='Blues', xticklabels=np.unique(y), yticklabels=np.unique(y))
    plt.title(f"Confusion Matrix (Fold {fold} - Validation After Tuning)")
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.show()
```

# --- 11. Rata-Rata Metrik untuk Train+Validation Setelah Tuning ---

```python
tuned_val_performance = {
    'accuracy': round(np.mean(accuracy_scores_tuned), 4),
    'f1_score': round(np.mean(f1_scores_tuned), 4),
    'precision': round(np.mean(precision_scores_tuned), 4),
    'recall': round(np.mean(recall_scores_tuned), 4)
}

print("\nTuned Model Performance on Train+Validation Set:")
for metric, value in tuned_val_performance.items():
    print(f"{metric.capitalize()}: {value:.4f}")
```

# --- 12. Evaluasi Model Setelah Tuning pada Test Set ---

```python
best_model.fit(X_train_val, y_train_val)  # Melatih ulang pada seluruh Train+Validation
y_pred_test_tuned = best_model.predict(X_test)

# Menghitung performa model pada test set setelah tuning
tuned_test_performance = {
    'accuracy': accuracy_score(y_test, y_pred_test_tuned),
    'precision': precision_score(y_test, y_pred_test_tuned, average='weighted'),
    'recall': recall_score(y_test, y_pred_test_tuned, average='weighted'),
    'f1_score': f1_score(y_test, y_pred_test_tuned, average='weighted')
}

print("\nTuned Model Performance on Test Set:")
print(f"Accuracy: {tuned_test_performance['accuracy']:.4f}")
print(f"Precision: {tuned_test_performance['precision']:.4f}")
print(f"Recall: {tuned_test_performance['recall']:.4f}")
print(f"F1 Score: {tuned_test_performance['f1_score']:.4f}")
```

# --- 13. Perbandingan Baseline vs Tuning ---

```python
# --- 13. Perbandingan Baseline vs Tuning ---
print("\n--- Perbandingan Baseline vs Tuning ---")

# Metrik evaluasi untuk model baseline pada Test Set
print("\nPerformance on Test Set (Baseline):")
print(f"Accuracy: {test_performance['accuracy']:.4f}")
print(f"Precision: {test_performance['precision']:.4f}")
print(f"Recall: {test_performance['recall']:.4f}")
print(f"F1 Score: {test_performance['f1_score']:.4f}")

# Metrik evaluasi untuk model setelah tuning pada Test Set
print("\nPerformance on Test Set (Tuned):")
print(f"Accuracy: {tuned_test_performance['accuracy']:.4f}")
print(f"Precision: {tuned_test_performance['precision']:.4f}")
print(f"Recall: {tuned_test_performance['recall']:.4f}")
print(f"F1 Score: {tuned_test_performance['f1_score']:.4f}")

# **Visualisasi Perbandingan Performa**
# Menyiapkan data untuk visualisasi
comparison = pd.DataFrame({
    'Baseline': [
        test_performance['accuracy'], 
        test_performance['precision'], 
        test_performance['recall'], 
        test_performance['f1_score']
    ],
    'Tuned': [
        tuned_test_performance['accuracy'], 
        tuned_test_performance['precision'], 
        tuned_test_performance['recall'], 
        tuned_test_performance['f1_score']
    ]
}, index=['Accuracy', 'Precision', 'Recall', 'F1 Score'])

# Membuat bar plot untuk perbandingan
comparison.plot(kind='bar', figsize=(10, 6), color=['blue', 'green'])

# Menambahkan judul, label, dan grid
plt.title('Performance Comparison: Baseline vs After Tuning')
plt.ylabel('Score')
plt.legend(loc='lower right')
plt.grid(axis='y')

# Menampilkan bar plot
plt.show()
```
